```python
import pandas as pd
```

```python
credit=pd.read_excel('/content/ido.xlsx')
```

```python
credit
```

```python
import numpy as np
np.random.seed(seed=1)
a = np.random.randint(2, size=len(credit))
```

```python
train = credit[a == 1] # ข้อมูล ฝึก
train.shape
test = credit[a == 0] # ข้อสอบ
test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train = train.iloc[:,:-1]
y_train = train.iloc[:,-1]
X_test = test.iloc[:,:-1]
y_test = test.iloc[:,-1]
```

```python
from sklearn.metrics import accuracy_score
```

```python
features = []
tresholds = []
for random in range(1000):
    DtreeHW2_1 = DecisionTreeClassifier(criterion="entropy", splitter='best', random_state=random)
    DtreeHW2_1.fit(X_train,y_train)
    features.append(DtreeHW2_1.tree_.feature[0])
    tresholds.append(DtreeHW2_1.tree_.threshold[0])
```

```python
y_predict = DtreeHW2_1.predict(X_test)
```

```python
accuracy_score(y_test, y_predict)
```
